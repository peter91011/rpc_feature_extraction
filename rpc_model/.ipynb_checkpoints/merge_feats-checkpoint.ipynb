```python
for date in ref_dates:
    print('working on date '+date)
    # read header
    file = [x for x in files if x[-14:-4] == date and 'header' in x][0]
    df_feat_all = pd.read_csv(file)
    
    df_feat_all = df_feat_all[(~pd.isnull(df_feat_all['catalog_item_id']))&(~pd.isnull(df_feat_all['seller_id']))]
    df_feat_all['catalog_item_id'] = df_feat_all['catalog_item_id'].astype('int')
    df_feat_all['catalog_item_id'] = df_feat_all['catalog_item_id'].astype('str')
    df_feat_all['seller_id'] = df_feat_all['seller_id'].astype('int')
    df_feat_all['seller_id'] = df_feat_all['seller_id'].astype('str')
    df_feat_all.drop_duplicates(subset = 'adid', inplace=True)
    
    # drop off duplicates catalog_item_id, seller_id
    
#     # read targets
#     file = 'sem_l_'+date+'.csv'
#     df_tmp = pd.read_csv(file, index_col=0)
#     df_feat_all = df_tmp.merge(df_feat_all, left_on = 'catalog_item_id', right_on = 'catalog_item_id', how='left')
    
    # read sem coop features
    files_to_read = [x for x in files if x[-14:-4] == date and 'semcoop' in x]
    
    # left join to ensure item coverage is based on target coverage
    for file in files_to_read:
        print('merge sem coop features')
        df_tmp = pd.read_csv(file)
        df_tmp = df_tmp[(~pd.isnull(df_tmp['catalog_item_id']))&(~pd.isnull(df_tmp['seller_id']))]
        df_tmp['catalog_item_id'] = df_tmp['catalog_item_id'].astype('int')
        df_tmp['catalog_item_id'] = df_tmp['catalog_item_id'].astype('str')
        df_tmp['seller_id'] = df_tmp['seller_id'].astype('int')
        df_tmp['seller_id'] = df_tmp['seller_id'].astype('str')
        df_tmp.drop_duplicates(subset = 'adid', inplace=True)
        df_feat_all = df_feat_all.merge(df_tmp, on=['catalog_item_id', 'seller_id', 'source_id', 'adid', 'is_mobile'], how='left')
        del df_tmp
        
    # read catalog feature
    files_to_read = [x for x in files if x[-14:-4] == date and 'catalog' in x]
    for file in files_to_read:
        print('merge catalog features')
        df_tmp = pd.read_csv(file)
        df_tmp = df_tmp[(~pd.isnull(df_tmp['catalog_item_id']))&(~pd.isnull(df_tmp['seller_id']))]
        df_tmp['catalog_item_id'] = df_tmp['catalog_item_id'].astype('int')
        df_tmp['catalog_item_id'] = df_tmp['catalog_item_id'].astype('str')
        df_tmp['seller_id'] = df_tmp['seller_id'].astype('int')
        df_tmp['seller_id'] = df_tmp['seller_id'].astype('str')
        df_tmp.drop_duplicates(subset = ['catalog_item_id','seller_id'], inplace=True)
        df_feat_all = df_feat_all.merge(df_tmp, on=['catalog_item_id', 'seller_id'], how='left')
        del df_tmp
        
    # read site features
    files_to_read = [x for x in files if x[-14:-4] == date and 'site' in x]
    for file in files_to_read:
        print('merge site features')
        df_tmp = pd.read_csv(file)
        df_tmp = df_tmp[(~pd.isnull(df_tmp['catalog_item_id']))&(~pd.isnull(df_tmp['seller_id']))]
        df_tmp['catalog_item_id'] = df_tmp['catalog_item_id'].astype('int')
        df_tmp['catalog_item_id'] = df_tmp['catalog_item_id'].astype('str')
        df_tmp['seller_id'] = df_tmp['seller_id'].astype('int')
        df_tmp['seller_id'] = df_tmp['seller_id'].astype('str')
        df_tmp.drop_duplicates(subset = ['catalog_item_id','seller_id'], inplace=True)
        df_feat_all = df_feat_all.merge(df_tmp, on=['catalog_item_id', 'seller_id'], how='left')
        del df_tmp
        
    # read bounce features
    files_to_read = [x for x in files if x[-14:-4] == date and 'bounce' in x]
    for file in files_to_read:
        print('merge bounce features')
        df_tmp = pd.read_csv(file)
        df_tmp = df_tmp[(~pd.isnull(df_tmp['catalog_item_id']))&(~pd.isnull(df_tmp['seller_id']))]
        df_tmp['catalog_item_id'] = df_tmp['catalog_item_id'].astype('int')
        df_tmp['catalog_item_id'] = df_tmp['catalog_item_id'].astype('str')
        df_tmp['seller_id'] = df_tmp['seller_id'].astype('int')
        df_tmp['seller_id'] = df_tmp['seller_id'].astype('str')
        df_tmp.drop_duplicates(subset = ['catalog_item_id','seller_id'], inplace=True)
        df_feat_all = df_feat_all.merge(df_tmp, on=['catalog_item_id', 'seller_id'], how='left')
        del df_tmp

    # calculate same_brand and same_seller weighted price
    for i in ['uber_curr_item_price_w1','uber_avg_overall_rating_w1']:
        df_feat_all[i] = df_feat_all[i].apply(lambda x: pd.to_numeric(x, errors='coerce'))
    print('dtype changes')
    
    df_feat_all['weight_price_cart_w1'] = df_feat_all['site_total_add_to_carts_w1']*(df_feat_all['uber_curr_item_price_w1'] - df_feat_all['uber_price_change_value_w1'])
    df_feat_all['weight_price_cart_w2'] = df_feat_all['site_total_add_to_carts_w2']*(df_feat_all['uber_curr_item_price_w1'] - df_feat_all['uber_price_change_value_w1'])
    df_feat_all['weight_price_orders_w1'] = df_feat_all['site_total_orders_w1']*(df_feat_all['uber_curr_item_price_w1'] - df_feat_all['uber_price_change_value_w1'])
    df_feat_all['weight_price_orders_w2'] = df_feat_all['site_total_orders_w2']*(df_feat_all['uber_curr_item_price_w1'] - df_feat_all['uber_price_change_value_w1'])
    brand_price = df_feat_all.groupby('brand_nm').agg({'weight_price_cart_w1':'sum','weight_price_cart_w2':'sum',
                                                      'site_total_add_to_carts_w1':'sum','site_total_add_to_carts_w2':'sum',
                                                      'weight_price_orders_w1':'sum','weight_price_orders_w2':'sum',
                                                      'site_total_orders_w1':'sum','site_total_orders_w2':'sum'}).reset_index()
    brand_price['uber_weight_price_cart_w1_brand'] = brand_price['weight_price_cart_w1']/brand_price['site_total_add_to_carts_w1']
    brand_price['uber_weight_price_cart_w2_brand'] = brand_price['weight_price_cart_w2']/brand_price['site_total_add_to_carts_w2']
    brand_price['uber_weight_price_orders_w1_brand'] = brand_price['weight_price_orders_w1']/brand_price['site_total_orders_w1']
    brand_price['uber_weight_price_orders_w2_brand'] = brand_price['weight_price_orders_w2']/brand_price['site_total_orders_w2']
    df_feat_all = df_feat_all.merge(brand_price[['brand_nm','uber_weight_price_cart_w1_brand','uber_weight_price_cart_w2_brand','uber_weight_price_orders_w1_brand','uber_weight_price_orders_w2_brand']],
                                   on = 'brand_nm',how = 'left')
    del brand_price
    print('del brand weight price')
    
    seller_price = df_feat_all.groupby('seller_id').agg({'weight_price_cart_w1':'sum','weight_price_cart_w2':'sum',
                                                      'site_total_add_to_carts_w1':'sum','site_total_add_to_carts_w2':'sum',
                                                      'weight_price_orders_w1':'sum','weight_price_orders_w2':'sum',
                                                      'site_total_orders_w1':'sum','site_total_orders_w2':'sum'}).reset_index()
    seller_price['uber_weight_price_cart_w1_seller'] = seller_price['weight_price_cart_w1']/seller_price['site_total_add_to_carts_w1']
    seller_price['uber_weight_price_cart_w2_seller'] = seller_price['weight_price_cart_w2']/seller_price['site_total_add_to_carts_w2']
    seller_price['uber_weight_price_orders_w1_seller'] = seller_price['weight_price_orders_w1']/seller_price['site_total_orders_w1']
    seller_price['uber_weight_price_orders_w2_seller'] = seller_price['weight_price_orders_w2']/seller_price['site_total_orders_w2']
    df_feat_all = df_feat_all.merge(seller_price[['seller_id','uber_weight_price_cart_w1_seller','uber_weight_price_cart_w2_seller','uber_weight_price_orders_w1_seller','uber_weight_price_orders_w2_seller']],
                                   on = 'seller_id',how = 'left')
    del seller_price
    print('del seller weight price')
    
    df_feat_all = df_feat_all.drop(['weight_price_cart_w1','weight_price_cart_w2','weight_price_orders_w1','weight_price_orders_w2'],axis=1)
    
    upload_csv_to_cloud_storage(df_feat_all, gcp_directoty_full+'df_feat_all_coop_item_'+date+'.csv')
    
    #cluster similar items for the same seller, from adid level to item_variant+seller+is_mobile+source_id level
    click_threshold = 150/7
    df_feat_all['f_threshold'] = df_feat_all['sem_clicks_l'].map(lambda x: 1 if x>=click_threshold else 0)
    df_feat_all1 = df_feat_all[df_feat_all['f_threshold']==1]
    df_feat_all2 = df_feat_all[df_feat_all['f_threshold']==0]
    
    agg_dic = {}
    agg_dic['catalog_item_id'] = lambda x:'__'.join(x)
    for i in df_feat_all.columns:
        if i.startswith('sem') and any(ext in i for ext in ['impressions','clicks','adspend','qty','orders','gmv_revenue_adjustment','cost','margin','estimated_cp']):
            agg_dic[i] = lambda x:x[x.notnull()].sum()
        if i.startswith('uber') and any(ext in i for ext in ['curr_item_price','price_change_desc','avg_overall_rating','avg_cost','price_change_value','avg_margin']):
            agg_dic[i] = lambda x:x[x.notnull()].mean()
        if i.startswith('uber') and any(ext in i for ext in ['num_appr_reviews','weight']):
            agg_dic[i] = lambda x:x[x.notnull()].sum()
        if i.startswith('site') and any(ext in i for ext in ['total_page_views','total_add_to_carts','total_revenue','margin','total_orders']):
            agg_dic[i] = lambda x:x[x.notnull()].sum()
        if i.startswith('bc') and any(ext in i for ext in ['bounces','visits','sem_visits','sem_bounces','total_orders']):
            agg_dic[i] = lambda x:x[x.notnull()].sum()
    aa = df_feat_all2.groupby(['id_unique_variant','seller_id','source_id','is_mobile']).agg(agg_dic).reset_index()
    
    aa['bc_bounce_rate_w1'] = aa['bc_bounces_w1']/aa['bc_visits_w1']
    aa['bc_bounce_rate_w2'] = aa['bc_bounces_w2']/aa['bc_visits_w2']
    aa['bc_sem_bounce_rate_w1'] = aa['bc_sem_bounces_w1']/aa['bc_sem_visits_w1']
    aa['bc_sem_bounce_rate_w2'] = aa['bc_sem_bounces_w2']/aa['bc_sem_visits_w2']
    aa['site_atcpv_w1'] = aa['site_total_add_to_carts_w1']/aa['site_total_page_views_w1']
    aa['site_atcpv_w2'] = aa['site_total_add_to_carts_w2']/aa['site_total_page_views_w2']
    aa['site_convrt_w1'] = aa['site_total_orders_w1']/aa['site_total_page_views_w1']
    aa['site_convrt_w2'] = aa['site_total_orders_w2']/aa['site_total_page_views_w2']
    aa['site_mpv_w1'] = aa['site_margin_w1']/aa['site_total_page_views_w1']
    aa['site_mpv_w2'] = aa['site_margin_w2']/aa['site_total_page_views_w2']
    aa['site_rpv_w1'] = aa['site_total_revenue_w1']/aa['site_total_page_views_w1']
    aa['site_rpv_w2'] = aa['site_total_revenue_w2']/aa['site_total_page_views_w2']
    aa['uber_price_change_ratio_w1'] = aa['uber_curr_item_price_w1']/(aa['uber_curr_item_price_w1'] - aa['uber_price_change_value_w1'])
    for name in [i.split('_')[-1] for i in df_feat_all.columns if i.startswith('sem')]:
        aa['sem_convrt_brand_'+name] = aa['sem_orders_brand_'+name]/aa['sem_clicks_'+name]
        aa['sem_convrt_seller_'+name] = aa['sem_orders_seller_'+name]/aa['sem_clicks_'+name]
        aa['sem_cpc_'+name] = aa['sem_adspend_'+name]/aa['sem_clicks_'+name]
        aa['sem_ctr_'+name] = aa['sem_clicks_'+name]/aa['sem_impressions_'+name]
        aa['sem_mpc_brand_'+name] = (aa['sem_gmv_revenue_adjustment_brand_'+name]-aa['sem_cost_'+name])/aa['sem_clicks_'+name]
        aa['sem_mpc_seller_'+name] = (aa['sem_gmv_revenue_adjustment_seller_'+name]-aa['sem_cost_'+name])/aa['sem_clicks_'+name]
        aa['sem_mpo_brand_'+name] = (aa['sem_gmv_revenue_adjustment_brand_'+name]-aa['sem_cost_'+name])/aa['sem_orders_brand_'+name]
        aa['sem_mpo_seller_'+name] = (aa['sem_gmv_revenue_adjustment_seller_'+name]-aa['sem_cost_'+name])/aa['sem_orders_seller_'+name]
        aa['sem_roas_brand_'+name] = aa['sem_gmv_revenue_adjustment_brand_'+name]/aa['sem_adspend_'+name]
        aa['sem_roas_seller_'+name] = aa['sem_gmv_revenue_adjustment_seller_'+name]/aa['sem_adspend_'+name]
        aa['sem_rpc_brand_'+name] = aa['sem_gmv_revenue_adjustment_brand_'+name]/aa['sem_clicks_'+name]
        aa['sem_rpc_seller_'+name] = aa['sem_gmv_revenue_adjustment_seller_'+name]/aa['sem_clicks_'+name]
    df_feat_all = pd.concat([df_feat_all1[aa.columns], aa])
    
    upload_csv_to_cloud_storage(df_feat_all, gcp_directoty_full+'df_feat_all_coop_itemcluster_'+date+'.csv')
```
